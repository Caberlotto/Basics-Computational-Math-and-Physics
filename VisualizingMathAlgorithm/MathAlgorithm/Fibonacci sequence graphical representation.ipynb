### Fibonacci sequence graphical representation

This notebook consists of the graphic visualization on a Cartesian plane of the fibonacci sequence.
For the best graphic representation from sequence it was chosen to use the first 4 numbers.
The fibonacci sequence algorithm was written using a recursive function.

```python
def fibo(num):
    '''Fibonacci sequence algorithm'''
    if num == 0:
        return 0
    elif num == 1:
        return 1
    else:
        return fibo(num -2) + fibo(num - 1)
```

```python
x_number, y_number, num = [],[],24

for i in range(num):
    x_number.append(i+1) # Number
    y_number.append(fibo(i)) # Values
```

```python
import matplotlib.pyplot as plt
plt.plot(x_number, y_number, marker='o')
plt.xlabel('x-coordinate: Sequence number')
plt.ylabel('y-coordinate: Value')
plt.title('Fibonacci sequence graphical representation')
plt.show()
```
