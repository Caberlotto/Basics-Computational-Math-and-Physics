```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def print_f(x,y):
    
    plt.plot(x,y,'green',linewidth=2)
    plt.axis('square')
    plt.axis('off')
    
```

```python
k = 4.2
t = np.linspace(0,8*np.pi,1000)
x = np.cos(k*t) * np.cos(t)
y = np.cos(k*t) * np.sin(t)

print_f(x,y)
plt.show()
```
