## Drawing the Trajectory

```python
from matplotlib import pyplot as plt
import math
```

```python
def d_graph(x, y):
    plt.plot(x, y)
    plt.xlabel('x-coordinate: meters')
    plt.ylabel('y-coordinate: height reached')
    plt.title('Projectile motion of a ball')
```

```python
def frange(start, final, intervall):
    
    numbers = []
    
    while start < final:
        numbers.append(start)
        start = start + intervall
    
    return numbers
```

```python
def draw_trajectory(u, theta):
    
    theta = math.radians(theta) 
    #convert the angle (theta) from degrees to radians using the math.radians() function
    
    G = 9.8
    #create a label (G) to refer to the value of acceleration due to gravity, 9.8 m/s^2
    
    t_flight = 2*u*math.sin(theta)/G
    # calculate the time of flight
    
    intervals = frange(0, t_flight, 0.001)
    #values for start, final, and increment set to 0, t_flight, and 0.001
    
    x, y = [],[]

    for t in intervals:
        
        x.append(u*math.cos(theta)*t)
        y.append(u*math.sin(theta)*t - 0.5*G*t*t)
        # calculate the x- and y-coordinates for the trajectory at each of the time instants and store them in two separate lists, x and y
        # use the formulas for the distances Sx and Sy to calculate the coordinates
        
    d_graph(x,y)
```

```python
try:
    u = float(input('Initial velocity in m/s: '))
    theta = float(input('Angle of projection in degrees: '))
except ValueError:
    print("Invalid input")
else:
    draw_trajectory(u,theta)
    plt.show()
```

Comparing the Trajectory at Different Initial Velocities

```python
u_list = [20, 40, 60,80,100]
try:
    theta = float(input('Angle of projection in degrees: '))
except ValueError:
    print("Invalid input")
else:
    for u in u_list:
        draw_trajectory(u,theta)

    plt.legend([20, 40, 60,80,100])
    plt.show()
```
